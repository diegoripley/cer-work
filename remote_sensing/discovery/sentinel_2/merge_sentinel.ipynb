```python
import rasterio
```

```python
image_path = './sentinel2/GRANULE/L2A_T12UUB_A018686_20201003T184314/IMG_DATA/R10m'
```

```python
band_4 = rasterio.open(f'{image_path}/T12UUB_20201003T184129_B04_10m.jp2', driver='JP2OpenJPEG')
band_3 = rasterio.open(f'{image_path}/T12UUB_20201003T184129_B03_10m.jp2', driver='JP2OpenJPEG')
band_2 = rasterio.open(f'{image_path}/T12UUB_20201003T184129_B02_10m.jp2', driver='JP2OpenJPEG')
```

```python
band_4.width
```

```python
true_color = rasterio.open('T12UUB_20201003T184129_True.tiff', 'w', driver='Gtiff',
                          width=band_4.width, height=band_4.height,
                          count=3,
                          crs=band_4.crs,
                          transform=band_4.transform,
                          dtype=band_4.dtypes[0])
```

```python
true_color.write(band_2.read(1), 3) # blue
true_color.write(band_3.read(1), 2) # green
true_color.write(band_4.read(1), 1) # red
```

```python
true_color.close()
```
