```bash
# NVME Storage
output_data_directory=/home/jovyan/work/notebooks/CER/experiment_1/data

cd /data
# Files to process
#files=$(ls *.vrt)
files=`cat /data/tiles_process.txt`
for tileid in ${files}; do
    output_file="${output_data_directory}/${tileid}.tif"
    input_file="/data/${tileid}.vrt"
    
    if [ ! -f ${input_file} ]; then
        continue
        echo "${input_file} not found"
    fi
    
    echo "Processing ${input_file}"
    gdalwarp -t_srs EPSG:900913 \
        ${input_file} ${output_file} \
        -co TILED=YES -co COPY_SRC_OVERVIEWS=YES -co COMPRESS=DEFLATE \
        -multi \
        -wo NUM_THREADS=ALL_CPUS \
        -co ALPHA=YES
    #sed 's/\/data\///g' ${output_file} -i
done
```
